```python
import pandas as pd

data_dir = "../../../data/"

target = pd.read_csv(data_dir + "s&p_adjusted_target.csv").set_index(["date", "ticker"])
target.head()
```

```python
import matplotlib.pyplot as plt

target[["target", "s&p_change"]].groupby("date").median().plot(figsize=(12, 10))
plt.title("Market Trend Comparison between Target and S&P 500")
plt.xticks(rotation = 45)
plt.xlabel("Date")
plt.ylabel("% Change with Previous Week")
plt.savefig("market_trend.png", dpi = 300)
```

```python
target[["change_adj_s&p", "s&p_change"]].groupby("date").median().plot(figsize=(12, 10))
plt.title("Market Trend Comparison between Adjusted target and S&P 500")
plt.xticks(rotation = 45)
plt.xlabel("Date")
plt.ylabel("% Change with Previous Week")
plt.savefig("adjusted_market_trend.png", dpi = 300)
```
