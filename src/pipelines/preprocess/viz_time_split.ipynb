```python
from sklearn.model_selection import TimeSeriesSplit
import numpy as np
import pandas as pd
```

```python
train_dates = ['2023-10-02', '2023-10-03', '2023-10-04', '2023-10-05', '2023-10-06',
        '2023-10-09', '2023-10-10', '2023-10-11', '2023-10-12', '2023-10-13',
        '2023-10-16', '2023-10-17', '2023-10-18', '2023-10-19', '2023-10-20',
        '2023-10-24', '2023-10-25', '2023-10-26', '2023-10-27']
val_dates = ['2023-10-30', '2023-10-31', '2023-11-01', '2023-11-02', '2023-11-03']
test_dates = ['2023-11-06', '2023-11-07', '2023-11-08', '2023-11-09', '2023-11-10',
       '2023-11-13', '2023-11-14']

all_dates = train_dates + val_dates + test_dates
```

```python
time_split = TimeSeriesSplit(n_splits = 5, test_size = 1, gap = 5)

train_val_split = list(time_split.split(train_dates + val_dates))
```

```python
train_val_split
```

```python
df = pd.DataFrame({
	"Initial Padding": [0, 1, 2, 3, 4],
	"Train Dates": [10] * 5,
	"Data Leakage Padding": [5] * 5,
	"Val Dates": [1] * 5
})
```

```python
import matplotlib.pyplot as plt
df.plot(kind="barh", stacked = True, color=["gray", "orange", "gray", "blue"], figsize=(12, 6))
plt.subplots_adjust(right=0.75)
plt.xticks(ticks = [i for i in range(0, 21, 5)], labels = [all_dates[2+i] for i in range(0, 21, 5)])
plt.xlabel("Dates")
plt.ylabel("Splits")
plt.title("Cross Validation Splits for Train/Validation (5 splits)")
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.savefig("train_split_viz.png", dpi = 300)
```
